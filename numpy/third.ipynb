```python
import numpy as np
import matplotlib.pyplot as pit
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("=========Sales Analysis=========")
print("\n sales Data Shape: ", sales_data.shape)
print("\nSample data for first 3 restraunt: ", sales_data[:, 1:])
```

```python
# Total Sales per year
print(np.sum(sales_data, axis = 0))
yearly_sum = np.sum(sales_data[:,1:], axis = 0)
print(yearly_sum)

# Total Sales per Restraunt
total_sum = np.sum(sales_data[:,1:], axis = 1)
print(total_sum)

# Maximum Sales per Restraunt
yearly_max = np.max(sales_data[:,1:], axis = 1)
print(yearly_max)

# Maximum Sales per year
yearly_max = np.max(sales_data[:,1:], axis = 0)
print(yearly_max)

# Average Sales Per Restraunt
yearly_avg = np.mean(sales_data[:,1:], axis = 1)
print(yearly_avg)

# Cumilative sum
cumsum = np.cumsum(sales_data[:,1:], axis = 1)
print(cumsum)
```

```python
# Plotting the cumilative
pit.figure(figsize=(10,6))
pit.plot(np.mean(cumsum, axis = 0))
pit.title("Average cumulative sales across all Restraunts")
pit.xlabel("Years")
pit.ylabel("Sales")
pit.grid(True)
pit.show()
```
